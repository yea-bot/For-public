```python
import pandas as pd
from datetime import timedelta
import os

# 함수 정의
def change_sec(time):
    days, hours, minutes, seconds_milliseconds = time.split(':')
    seconds, milliseconds = seconds_milliseconds.split('.')
    time_delta = timedelta(days=int(days), hours=int(hours), minutes=int(minutes), seconds=int(seconds), milliseconds=int(milliseconds))
    return time_delta.total_seconds()

def process_file(file_path, new_directory_path):
    # Pandas를 사용하여 CSV 파일을 읽기   
    df_res = pd.read_csv(file_path)
    cell_name = os.path.basename(file_path).replace('_DC.csv', '')[:-4]

    # 최대 전압 설정
    max_V = 4.3 if cell_name[2:5] == '4.3' else 4.2
    print('max V:', max_V)
    print('cell name:', cell_name)

    # 유일한 Cycle_No. 값 가져오기
    cycle_numbers = df_res["Cycle_No."].unique()

    cycle_data = []

    # 각 사이클에 대해 반복 처리
    for cyc in cycle_numbers:
        cur_df = df_res[df_res['Cycle_No.'] == cyc]

        # 방전 스텝 번호 추출
        # 각 스텝별 "Current(A)" 값의 평균 계산
        mean_current_per_step = cur_df.groupby('Step_No.')['Current(A)'].mean()

        # 평균이 음수인 스텝 추출
        negative_current_steps = mean_current_per_step[mean_current_per_step < 0].index

        # 첫 번째 음수 평균 스텝 번호 찾기
        first_negative_step = negative_current_steps.min() if not negative_current_steps.empty else None

        # 필터링 조건 확인
        if first_negative_step is not None:
            # 그 스텝 이후의 모든 스텝 포함하여 필터링
            discharge_step_df = cur_df[cur_df['Step_No.'] >= first_negative_step]

            # 고유한 스텝 번호만 추출
            unique_discharge_steps = discharge_step_df['Step_No.'].unique()

            discharge_step = unique_discharge_steps.tolist()
            
            # print('extracted discharge steps:', discharge_step)
            
            # 방전 스텝의 인덱스 계산
            index = cur_df[cur_df['Step_No.'].isin(discharge_step)].index

            # 충전 및 방전 데이터프레임 생성
            chr_df = cur_df.copy()
            dis_df = cur_df.copy()

            # 충전 데이터프레임에서 방전 부분의 용량을 0으로 설정
            chr_df.loc[index, '|Q|(Ah)'] = 0
            # 방전 데이터프레임에서 충전 부분의 용량을 0으로 설정
            dis_df.loc[~dis_df.index.isin(index), '|Q|(Ah)'] = 0

            # 데이터를 리스트로 변환하여 수집
            discharge_capacity_in_Ah = dis_df['|Q|(Ah)'].tolist()
            charge_capacity_in_Ah = chr_df['|Q|(Ah)'].tolist()
            assert len(discharge_capacity_in_Ah) == len(charge_capacity_in_Ah), "Capacity length of dis & chr is not equal"

            # 시간을 초로 변환하여 수집
            timedata = cur_df['Test_Time(s)'].tolist()
            time_in_s = [change_sec(time) for time in timedata]

            # cycle 데이터를 딕셔너리로 저장
            cyc_dict = {
                'cycle_number': int(cyc),
                'step': cur_df['Step_No.'].tolist(),
                'time_in_s': time_in_s,
                'current_in_A': cur_df["Current(A)"].tolist(),
                'voltage_in_V': cur_df['Voltage(V)'].tolist(),
                'discharge_capacity_in_Ah': discharge_capacity_in_Ah,
                'charge_capacity_in_Ah': charge_capacity_in_Ah,
                'temperature_in_C': None,  # None으로 설정
                'internal_resistance_in_ohm': None  # None으로 설정
            }
            cycle_data.append(cyc_dict)

    # 결과를 저장할 데이터프레임 생성
    dataframes = {
        'cell_id': cell_name,
        'cycle_data': cycle_data,
        'form_factor': 'coin_cell',
        'anode_material': 'Li-metal',
        'cathode_material': 'NCM523',
        'electrolyte_material': None,
        'nominal_capacity_in_Ah': 0.0034,
        'depth_of_charge': 1.0,
        'depth_of_discharge': 1.0,
        'already_spent_cycles': int(cycle_numbers.min() - 1),
        'max_voltage_limit_in_V': max_V,
        'min_voltage_limit_in_V': 3.0,
        'max_current_limit_in_A': None,
        'min_current_limit_in_A': None,
        'reference': None,
        'description': None,
        'charge_protocol': None,
        'discharge_protocol': None,
        'relaxation_protocol': None,
        'short_circuit_cycle': len(cycle_data)-1
    }

    # 결과를 파일로 저장
    new_file_name = cell_name + '.pkl'
    new_file_path = os.path.join(new_directory_path, new_file_name)
    pd.to_pickle(dataframes, new_file_path)
    print(f'Success to save \'{new_file_path}\'')
    

for file_path in file_list:
    process_file(file_path, new_directory_path)
```
